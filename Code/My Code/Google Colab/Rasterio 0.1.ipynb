```python
with rio.open(r"/content/data/Images Output/sentinel2_2017-07-04_out.tiff", 'w',

              driver = ds.driver,
              height = ds.height,
              width = ds.width,
              count = ds.count,
              crs = ds.crs,
              transform = ds.transform,
              dtype = data.dtype,
              ) as dst:
  dst.write(data)
```
