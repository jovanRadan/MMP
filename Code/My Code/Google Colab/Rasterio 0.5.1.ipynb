```python
import rasterio as rio
import pandas as pd
import numpy as np
```

```python
img_path = r'/content/data/image/sentinel2_2017-07-04.tiff'
mask_path = r'/content/data/mask/EMSR207_06MADEIRA_02GRADING_MAP_v2_vector_mask.tiff'

img = rio.open(img_path)
mask = rio.open(mask_path)

array = img.read()
#num_bands = array.shape[0]
num_bands = 12

height = img.shape[0]
width = img.shape[1]

mask_data = mask.read(1)
mask_data = np.expand_dims(mask_data, axis = 0)

combined_array = np.concatenate((array, mask_data), axis=0)

df = pd.DataFrame(array.reshape([num_bands+1,-1]).T, columns=[f"band_{i+1}" for i in range(num_bands)]+['mask'])

if 'band_13' in df.columns:
  df = df.drop(columns=['band_13'])

column_names = df.columns

try:
  if df is not None:
    print(df)
    print(column_names)
    print(num_bands)
  else:
    print("Error: Invalid transformation matrix in image metadata")
except Exception as e:
  print(f"Error: {e}")
```
